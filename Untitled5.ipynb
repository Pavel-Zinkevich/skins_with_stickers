```python
base_url = 'https://steamcommunity.com/market/search/render/'

# –í—Å–µ —Ç–≤–æ–∏ —Ñ–∏–ª—å—Ç—Ä—ã, –≤–∫–ª—é—á–∞—è AK-47
params_template = {
    'appid': 730,
    'category_730_ItemSet[]': 'any',
    'category_730_ProPlayer[]': 'any',
    'category_730_StickerCapsule[]': 'any',
    'category_730_Tournament[]': 'any',
    'category_730_TournamentTeam[]': 'any',
    'category_730_Type[]': 'any',
    'category_730_Weapon[]': 'tag_weapon_usp_silencer',
    'category_730_Exterior[]': [
        'tag_WearCategory0', 'tag_WearCategory1', 'tag_WearCategory2',
        'tag_WearCategory3', 'tag_WearCategory4'
    ],
    'category_730_Quality[]': ['tag_normal', 'tag_tournament', 'tag_strange'],
    'count': 10,  # —Ä–µ–∞–ª—å–Ω–æ 10 —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –Ω–∞ —Å—Ç—Ä–∞–Ω–∏—Ü—É
}

names = []
prices = []
links = []
seen_items = set()

max_pages = 60  # –°–∫–æ–ª—å–∫–æ —Å—Ç—Ä–∞–Ω–∏—Ü –∑–∞–≥—Ä—É–∂–∞—Ç—å –º–∞–∫—Å–∏–º—É–º (10 * 20 = 200 —Å–∫–∏–Ω–æ–≤)

for page in range(max_pages):
    start = page * params_template['count']
    params = params_template.copy()
    params['start'] = start

    print(f"üîÑ –ó–∞–≥—Ä—É–∂–∞–µ–º —Å—Ç—Ä–∞–Ω–∏—Ü—É {page + 1} (start={start})")

    try:
        response = requests.get(base_url, headers=headers, params=params)
        if response.status_code == 429:
            print("‚ö†Ô∏è  –ü–æ–ª—É—á–µ–Ω 429. –°–ø–∏–º –ø–æ–¥–æ–ª—å—à–µ...")
            time.sleep(random.uniform(5, 10))
            continue

        if not response.ok:
            print(f"‚ùå –û—à–∏–±–∫–∞ {response.status_code}")
            break

        data_json = response.json()
        html_content = data_json.get('results_html', '')

        soup = BeautifulSoup(html_content, 'html.parser')
        elements = soup.find_all('a', class_='market_listing_row_link')

        if not elements:
            print("üö´ –†–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –±–æ–ª—å—à–µ –Ω–µ—Ç.")
            break

        for el in elements:
            name_tag = el.find(class_='market_listing_item_name')
            price_tag = el.find(class_='sale_price')
            link = el.get('href')

            if not name_tag or not link:
                continue

            name = name_tag.text.strip()
            if name in seen_items:
                continue  # –¥—É–±–ª–∏–∫–∞—Ç

            seen_items.add(name)
            price = price_tag.text.strip() if price_tag else 'N/A'

            names.append(name)
            prices.append(price)
            links.append(link)

        time.sleep(random.uniform(1.5, 3))

    except Exception as e:
        print(f"‚ùó –ò—Å–∫–ª—é—á–µ–Ω–∏–µ: {e}")
        break

# –°–æ–±–∏—Ä–∞–µ–º DataFrame
df = pd.DataFrame({
    'Item Name': names,
    'Sale Price': prices,
    'Link': links
})
```

```python
for link in df['Link']:
    print(link+'?filter=2015+2014')
```

```python
sticker_prices = []  # to store prices for each link
sticker = '2014'
for link in df['Link']:
    price_found = None

    for year in [sticker]:
        url = link + f'?filter={year}'
        print(f"Checking: {url}")

        try:
            response = requests.get(url, headers=headers)
            if response.status_code == 429:
                print("‚ö†Ô∏è Received 429 Too Many Requests. Sleeping...")
                time.sleep(random.uniform(5, 10))
                continue
            elif not response.ok:
                print(f"Error: {response.status_code} on {url}")
                continue

            soup = BeautifulSoup(response.text, 'html.parser')
            rare_stickers = soup.find_all(class_="market_listing_price market_listing_price_with_fee")

            if rare_stickers:
                price_found = rare_stickers[0].text.strip()
                break  # Stop checking after first match

        except Exception as e:
            print(f"Exception occurred: {e}")
            continue

        time.sleep(random.uniform(1.5, 3))  # avoid rate limit

    if not price_found:
        price_found = "Not found"

    print(price_found)
    sticker_prices.append(price_found)
    time.sleep(random.uniform(2, 4))  # final delay to avoid getting blocked

# Add result to DataFrame
df[sticker] = sticker_prices
```
